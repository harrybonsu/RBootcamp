```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Set x axis to numerical value for month
x_axis = np.arange(1,13,1)
x_axis
```

```python
# Avearge weather temp
points_F = [39, 42, 51, 62, 72, 82, 86, 84, 77, 65, 55, 44]
```

```python
# Convert to Celsius C = (F-32) * 0.56
points_C = [(x-32) * 0.56 for x in points_F]
points_C
```

```python
# Create a handle for each plot
farenheit, = plt.plot(x_axis, points_F, marker='s', color='red', label='Farenheit')
celsius, = plt.plot(x_axis, points_C, marker='o', color='blue', label='Celcius')
```

```python
# Set our legend to where the chart thinks is best
plt.legend(loc='best')
plt.show()
```

```python
# Create labels for the X and Y axis
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```
