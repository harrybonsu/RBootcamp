```python
# Import Dependencies
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create the X and Y axis lists
time = np.arange(0,130,10)
danger_drop_speeds = [9, 8, 90, 85, 80, 70, 70, 65, 55, 60, 70, 65, 50]
railgun_speeds = [75, 70, 60, 65, 60, 45, 55, 50, 40, 40, 35, 35, 30]
```

```python
danger_drop, = plt.plot(time,danger_drop_speeds, color='red', marker='x', label='danger drop')
railgun, = plt.plot(time,railgun_speeds, color='blue', marker='o', label='railgun speeds')
plt.show()
```

```python
# Add labels to X and Y axes :: Add title
plt.title("Coaster Speed Over Time")
plt.xlabel("Coaster Runtime")
plt.ylabel("Speed(MPH)")
```

```python
# Set the limits for the X and Y axes
plt.xlim(0,120)
plt.ylim(0,95)
```

```python
# Create a legend for the chart
plt.legend(loc='best')
```

```python
# Add in a grid for the chart
plt.grid()
```
